Để làm câu c ta in ra label dự đoán được. Tạo một cột dữ liệu mới để dánh giá label mới tạo, đánh số 0 là trường tư, số 1 là trường công (Vì nếu đảo ngẫu nhiên đánh số 1 là trường tư và số 0 là trường công thì precicion và recall thấp hơn) . Tạo một df để so sánh kết quả của kmeans

````
def converter(cluster):
    if cluster=='Yes':
        return 0
    else:
        return 1
````

````
df['Cluster'] = df['Private'].apply(converter)
````

````
colleges = df
labels = kmeans.labels_
colleges['labels'] = labels
colleges = colleges[['Cluster','labels']]
````

````
sortcolleges = colleges.sort_values(['Cluster'])
pd.set_option('display.max_rows', 1000)
sortcolleges
````

Cuối cùng in ra kết quả ma trận nhiễu loạn và điểm Precision-Recall-F1.
ta nhận thấy Precision và Recall đều cao tại lớp 0 tuy nhiên đạt độ chính xác thấp tại lớp 1. Số lượng bị đánh label 0 cao hơn 2/3 so với lớp 1.

````
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
````
